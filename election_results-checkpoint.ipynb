```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://results.eci.gov.in/PcResultGenJune2024/index.htm')
```

```python
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python
table=soup.find('table')
```

```python
head=table.find_all('th')
t_columns=[i.text.strip() for i in head]
del t_columns[4:]
print(t_columns)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=t_columns)
df
```

```python
row_detail=table.find_all('tr')
row_detail
print(len(row_detail))
```

```python
for row in row_detail[1:43]:
    row_data=row.find_all('td')
    each_data=[data.text.strip() for data in row_data]
    print(each_data)
    
    length=len(df)
    df.loc[length]=each_data
```
